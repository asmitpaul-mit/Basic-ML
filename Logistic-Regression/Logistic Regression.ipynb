```python
import numpy as np
import pandas as pd
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def startWeights(nFeatures):
    b = np.zeros((nFeatures, 1))
    return b

def CalculateLogLikelihood(X, y, b):
    h = sigmoid(np.dot(X, b))
    likelihood = np.sum(y*np.log(h) + (1 - y)*np.log(1 - h))
    return likelihood

def logisticRegression(X, y, b, learningRate, iterations):
    m = len(y)
    likelihoods = []
    
    for i in range(iterations):
        h = sigmoid(np.dot(X, b))
        gradient = np.dot(X.T, (y - h))/m
        b += learningRate*gradient
        
        likelihood = CalculateLogLikelihood(X, y, b)
        likelihoods.append(likelihood)
        
    return b, likelihoods
```

```python
trainData = pd.read_csv('C:\\Users\\ASMITPAUL\\Downloads\\train(1).csv')
```

```python
data = pd.DataFrame(trainData)
gender = []

for i in data['Sex']:
    if i == 'male':
        gender.append(1)
    else:
        gender.append(0)

data['Sex'] = gender
data['Bias'] = 1
data
```

```python
XTrain = data[['Pclass', 'Sex', 'Age', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare', 'Bias']]
yTrain = data['Survived'].values.reshape(-1, 1)

b = startWeights(XTrain.shape[1])
learningRate = 0.01
iterations = 10000

b, likelihoods = logisticRegression(XTrain, yTrain, b, learningRate, iterations)
```
