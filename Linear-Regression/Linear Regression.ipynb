```python
import numpy as np
import pandas as pd
```

```python
trainData = pd.read_csv('C:\\Users\\ASMITPAUL\\Downloads\\train.csv')
trainData
```

```python
Y = np.matrix(trainData.iloc[:, 10:])

inputData = trainData.iloc[:, 0:10]
inputData.insert(loc = 0, column = 'Ones' ,value = 1)

X = inputData.to_numpy()
X = np.matrix(X)

print(X)
print(Y)
```

```python
parameters = np.linalg.inv(X.T @ X) @ X.T @ Y
print(list(np.array(parameters)))
```

```python
testData = pd.read_csv('C:\\Users\\ASMITPAUL\\Downloads\\test.csv')
testData.insert(loc = 0, column = 'Ones', value = 1)
testData
```

```python
prediction = []

for row in range(testData.shape[0]):
    predictedPrice = 0
    
    for column in range(testData.shape[1] - 1):
        predictedPrice = predictedPrice + testData.iloc[row,column]*parameters[column]
        
    prediction.append(predictedPrice)
    
prediction = np.asarray(prediction).squeeze(2)
prediction = prediction.squeeze(1)
prediction
```

```python
meanSale = np.mean(testData['SalePrice'])

ssTotal = np.sum((testData['SalePrice'] - meanSale)**2)
ssResidual = np.sum((testData['SalePrice'] - prediction)**2)

r2 = 1 - (ssResidual/ssTotal)
r2
```
